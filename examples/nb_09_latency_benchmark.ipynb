This notebook runs the benchmark script `scripts/benchmark_latency.py`, reads the generated CSV, and visualizes results.

## Run benchmark

The script will create series, insert values, perform updates and reads, and save results to CSV and PNG files. Adjust parameters as needed.

```python
# Run a small benchmark (adjust parameters if you want larger runs)
import subprocess
import shlex
import os

out_csv = "examples/bench_results.csv"
plots_dir = "examples/bench_plots"
cmd = f"python scripts/benchmark_latency.py --rows 10 100 --series 1 3 --repeats 2 --out {out_csv} --plots {plots_dir}"
print(cmd)
subprocess.run(shlex.split(cmd), check=True)

print("Done. Results written to:", out_csv)
print("Plots in:", plots_dir)
```

## Load and plot results

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("examples/bench_results.csv")
df
```

```python
# Simple throughput plot for each test type
for test in df['test'].unique():
    sub = df[df['test'] == test]
    plt.figure(figsize=(8,4))
    for ns in sorted(sub['num_series'].unique()):
        sel = sub[sub['num_series'] == ns]
        plt.plot(sel['total_values'], sel['throughput_rows_per_sec'], marker='o', label=f'{ns} series')
    plt.title(f"{test} throughput")
    plt.xlabel('Total values')
    plt.ylabel('Rows / sec')
    plt.grid(alpha=0.3)
    plt.legend()
    plt.tight_layout()
    plt.show()
```
